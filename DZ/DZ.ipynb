```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./res.csv')
df
```

```python
cdf = pd.DataFrame(df.to_numpy()[:, 1:] / 9.4 * 1000, columns=['2.82e-5 bar 300v', '2.54e-5 bar 350v', '2.39e-5 bar 400v',\
                            '3.32e-4 bar 300v', '3.31e-4 bar 350v', '3.30e-4 bar 400v', \
                           '1.1mb 300v', '1.1mb 350v', '1.1mb 400v'])
```

```python
#cdf
```

```python
errdf=pd.DataFrame(cdf*0.001/df.to_numpy()[:, 1:], columns=['2.82e-5 bar 300v', '2.54e-5 bar 350v', '2.39e-5 bar 400v',\
                            '3.32e-4 bar 300v', '3.31e-4 bar 350v', '3.30e-4 bar 400v', \
                           '1.1mb 300v', '1.1mb 350v', '1.1mb 400v'])
errdf
```

```python
upper = []
errupper = []
for i in cdf.to_numpy().T:
    x=df['v подаём (вольт)'][df['v подаём (вольт)']>=20][pd.notnull(i)[df['v подаём (вольт)']>=20]]
    y=i[df['v подаём (вольт)']>=20][pd.notnull(i)[df['v подаём (вольт)']>=20]]
    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y)[0]
    errm = 2*np.sqrt((1/(len(x)-2))*((np.mean(y**2)-np.mean(y)**2)/(np.mean(x**2)-np.mean(x)**2)-m**2))
    errc = np.sqrt((np.mean(x**2)-np.mean(x)**2) + np.mean(x)**2)*errm
    errupper.append([errm.copy(), errc.copy()])
    upper.append([m.copy(), c.copy()])
upper, errupper
```

```python
lower = []
errlower = []
for i in cdf.to_numpy().T:
    x=df['v подаём (вольт)'][df['v подаём (вольт)']<=-20][pd.notnull(i)[df['v подаём (вольт)']<=-20]]
    y=i[df['v подаём (вольт)']<=-20][pd.notnull(i)[df['v подаём (вольт)']<=-20]]
    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y)[0]
    errm = 2*np.sqrt((1/(len(x)-2))*((np.mean(y**2)-np.mean(y)**2)/(np.mean(x**2)-np.mean(x)**2)-m**2))
    errc = np.sqrt((np.mean(x**2)-np.mean(x)**2) + np.mean(x)**2)*errm
    errlower.append([errm.copy(), errc.copy()])
    lower.append([m.copy(), c.copy()])
lower, errlower
```

```python
fig, ax = plt.subplots(3,3, figsize=(20, 20))
ax = ax.ravel()
for ID, i in enumerate(ax):
    i.errorbar(df['v подаём (вольт)'][pd.notnull(cdf.to_numpy()[:, ID])],\
               cdf.to_numpy()[:, ID][pd.notnull(cdf.to_numpy()[:, ID])],\
               xerr=[0.1 for _ in np.arange(len(df['v подаём (вольт)'][pd.notnull(cdf.to_numpy()[:, ID])]))],\
              yerr=errdf.to_numpy()[:, ID][pd.notnull(cdf.to_numpy()[:, ID])])
    
    i.plot(np.array([0, 80]), np.array([0,80])*upper[ID][0]+upper[ID][1])
    i.plot(np.array([0, -80]), np.array([0,-80])*lower[ID][0]+lower[ID][1])
    #i.plot(np.array([10, -10]), np.array([10,-10])*steep[ID][0]+steep[ID][1])

    
    i.annotate('I_{+2}', xy=(0, upper[ID][1]),  xycoords='data',
            xytext=(-30, upper[ID][1]*1.1), textcoords = 'data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='left', verticalalignment='top',
            )
    i.annotate('I_{+1}', xy=(0, lower[ID][1]),  xycoords='data',
            xytext=(20, lower[ID][1]*1.1), textcoords = 'data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='left', verticalalignment='top',
            )
    
    i.axvline(linewidth=2, color='r')
    i.axhline(linewidth=2, color='r')
    i.set_title(cdf.columns[ID])
    i.set_xlim(-80, 80)
    i.set_ylim(np.round(cdf.to_numpy()[:, ID][pd.notnull(cdf.to_numpy()[:, ID])].min(), -1),\
                        np.round(cdf.to_numpy()[:, ID][pd.notnull(cdf.to_numpy()[:, ID])].max(), -1))
    i.set_xticks(np.arange(-80, 85, 10))
    i.set_yticks(np.arange(np.round(cdf.to_numpy()[:, ID][pd.notnull(cdf.to_numpy()[:, ID])].min(), -1),\
                           cdf.to_numpy()[:, ID][pd.notnull(cdf.to_numpy()[:, ID])].max()+10,\
                           10))
    i.grid()
    i.set_ylabel('current [mcA]', fontsize='large')
    i.set_xlabel('Voltage [V]', fontsize='large')
ax[6].set_yticks(np.arange(-3, 3.5, 0.5))
ax[6].set_ylim(-3, 3)
ax[7].set_yticks(np.arange(-10, 9, 1))
ax[7].set_ylim(-10, 8)
ax[8].set_yticks(np.arange(-20, 19, 2))
ax[8].set_ylim(-20, 18)
ax[3].set_yticks(np.arange(-40, 45, 5))
plt.savefig('./nineinone.pdf')
```

```python
plt.figure(figsize=(15, 15))
for ID, i in enumerate(cdf.to_numpy().T):
    plt.plot(df['v подаём (вольт)'][pd.notnull(i)], i[pd.notnull(i)], label=cdf.columns[ID])
#plt.plot([0,0], [110, -110], label='Oy')
#plt.plot([-80, 80], [0, 0], label='Ox')
plt.axvline(linewidth=2, color='r')
plt.axhline(linewidth=2, color='r')
plt.legend(fontsize='x-large')
plt.ylim(-110, 110)
plt.xlim(-80, 80)
plt.yticks(np.arange(-110, 115, 5))
plt.xticks(np.arange(-80, 85, 5))
plt.grid()
plt.ylabel('current [mcA]', fontsize='x-large')
plt.xlabel('Voltage [V]', fontsize='x-large')
plt.title('Current vs Voltage', fontsize='x-large')
plt.xticks(fontsize='large')
plt.yticks(fontsize='large')
plt.savefig('./allinone.pdf')
```
